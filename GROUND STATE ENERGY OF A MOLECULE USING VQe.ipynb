```python
from pennylane import numpy as np

symbols = ["H", "O", "H"]
coordinates = np.array([-0.0399, -0.0038, 0.0, 1.5780, 0.8540, 0.0, 2.7909, -0.5159, 0.0])
```

```python
import pennylane as qml

H, qubits = qml.qchem.molecular_hamiltonian(symbols, coordinates)

print("Number of qubits = ", qubits)
print("The Hamiltonian is ", H)
```

```python
electrons = 10
S2 = qml.qchem.spin2(electrons, qubits)
print(S2)
```

```python
hf = qml.qchem.hf_state(electrons, qubits)
print(hf)
```

```python
singles, doubles = qml.qchem.excitations(electrons, qubits, delta_sz=0)
print(singles)
print(doubles)
```

```python
def circuit(params, wires):
    qml.AllSinglesDoubles(params, wires, hf, singles, doubles)
```

```python
dev = qml.device("default.qubit", wires=qubits)
```

```python
@qml.qnode(dev)
def cost_fn(params):
    circuit(params, wires=range(qubits))
    return qml.expval(H)
```

```python
@qml.qnode(dev)
def S2_exp_value(params):
    circuit(params, wires=range(qubits))
    return qml.expval(S2)
```

```python
def total_spin(params):
    return -0.5 + np.sqrt(1 / 4 + S2_exp_value(params))
```

```python
opt = qml.GradientDescentOptimizer(stepsize=0.8)
np.random.seed(0)  # for reproducibility
theta = np.random.normal(0, np.pi, len(singles) + len(doubles), requires_grad=True)
print(theta)
```

```python
energy = [cost_fn(theta)]
angle = [theta]
max_iterations = 20
conv_tol = 1e-06

for n in range(max_iterations):

    theta, prev_energy = opt.step_and_cost(cost_fn, theta)

    energy.append(cost_fn(theta))
    spin = total_spin(theta)
    angle.append(theta)

    conv = np.abs(energy[-1] - prev_energy)

    if n % 2 == 0:
         print(f"Step = {n},Energy = {energy[-1]:.8f} Ha, S = {spin:.4f}")

    if conv <= conv_tol:
        break

print("\n" f"Final value of the ground-state energy = {energy[-1]:.8f} Ha")
print("\n" f"Optimal value of the circuit parameters = {theta}")
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
fig.set_figheight(5)
fig.set_figwidth(12)

# Full configuration interaction (FCI) energy computed classically
E_fci = -73.69851000

# Add energy plot on column 1
ax1 = fig.add_subplot(121)
ax1.plot(range(n + 2), energy, "go", ls="dashed")
ax1.plot(range(n + 2), np.full(n + 2, E_fci), color="red")
ax1.set_xlabel("Optimization step", fontsize=13)
ax1.set_ylabel("Energy (Hartree)", fontsize=13)
ax1.text(0.5,-67.85835195, r"$E_\mathrm{HF}$", fontsize=15)
ax1.text(0, -73.69851000, r"$E_\mathrm{FCI}$", fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# # Add angle plot on column 2
# ax2 = fig.add_subplot(122)
# ax2.plot(range(n + 2), angle, "go", ls="dashed")
# ax2.set_xlabel("Optimization step", fontsize=13)
# ax2.set_ylabel("Gate parameter $\\theta$ (rad)", fontsize=13)
# plt.xticks(fontsize=12)
# plt.yticks(fontsize=12)

plt.subplots_adjust(wspace=0.3, bottom=0.2)
plt.show()
```
